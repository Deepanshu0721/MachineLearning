__Train/Test split__

```python
X = df.drop(["target"], axis=1)
X.head()
```

```python
y = df["target"]
y.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler().fit(X)
X = scaler.transform(X)

X = pd.DataFrame(X)
X.columns = ['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach',
             'exang', 'oldpeak', 'slope', 'ca', 'thal']
X.head()
```

```python
#from sklearn.cross_validation import train_test_split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=1)
```

__Naive Bayes Classification__

```python
from sklearn.naive_bayes import GaussianNB
nbc = GaussianNB()
nbc.fit(X_train,y_train)
```

```python
y_pred = nbc.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print("The classification report is as follows...\n")
print(classification_report(y_pred,y_test))
```

```python
cm=np.array(confusion_matrix(y_test,y_pred))
confusion = pd.DataFrame(cm, index=["Heart_Disease", "Healthy"],
                         columns=["predicted_Heart_Disease", "predicted_healthy"])
confusion
```

```python
fig, ax =plt.subplots(figsize=(4,4))
plt.title("Confusion Matrix")
sns.heatmap(cm, mask=np.zeros_like(cm, dtype=np.bool), cmap=sns.diverging_palette(220, 10, as_cmap=True),
            square=True, ax=ax, annot=True,linewidths=3)
plt.show()
```

```python
accuracy = round((cm[0,0]+cm[1,1])/cm.sum(),3)
print('Accuracy = {:0.2f}%.'.format(100 * accuracy))
```

```python
cost= round((cm[0,0]* 0)+(cm[1,0]* 100)+ (cm[1,1]* -1)+ (cm[0,1]*1))
print("Misclassification Cost of the above model is",cost,".")
```
